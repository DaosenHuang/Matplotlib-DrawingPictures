```python
"""使用scatter()绘制散点图"""
import matplotlib as mpl
import numpy as np
import matplotlib.pyplot as plt
mpl.rcParams['font.sans-serif'] = ['SimHei']  # 中文字体支持

# 添加画布

x_values = [150,85,150,145,130]
y_values = [140,85,155,50,150]

plt.scatter(x_values, y_values,marker='^' ,s=50,color='r',label="客机")

# 设置图表标题并给坐标轴加上标签
plt.title('飞行记录数据', fontsize=18)
plt.grid("on") # 打开或者关闭栅格
plt.xlim(0,160) # x轴取值范围
plt.ylim(0,160) # y轴取值范围
plt.xlabel('横坐标 (km)', fontsize=14)
plt.ylabel('纵坐标 (km)', fontsize=14)
plt.legend()

# 设置刻度标记的大小
plt.tick_params(axis='both', which='major', labelsize=8)
for a, b in zip(x_values, y_values):  
    plt.text(a, b, (a,b),ha='left', va='top', fontsize=10) 
plt.show()
```
