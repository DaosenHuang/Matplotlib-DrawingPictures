```python
import matplotlib.pyplot as plt
import numpy as np
ax = plt.subplot(222)#  ax相当于画板中的子图
plt.subplot(223)
ax.plot([1,2])
plt.plot([2,1])
```

```python
fig, axes = plt.subplots(2,2)
print("画板：",fig)
print("子图：",axes)
plt.plot([1,3]) # 默认用最后一个子图作为当前图
```

```python
plt.style.use('seaborn')
x = np.linspace(0,10,100)# 0到10之间等分为100份
y1 = np.sin(x)
y2 = np.cos(x)
plt.plot(x,y1,label="y1=sin(x)")# 图形设置标签
plt.plot(x,y2,label="y2=cos(x)")
plt.title("Demo")# 标题
plt.xlabel('x')
plt.ylabel('The Value of y')# 横纵坐标的标签
plt.axis([0,10,-1,1])# 统一给横纵坐标设置坐标范围
plt.legend(loc=1)# 表示label显示
plt.xticks([0,3.1416,6.2832,9.4248],["0",'$\pi$','$2\pi$','$3\pi$'])# \用来转义
plt.yticks([-1,0,1])


         
```

```python
# 绘制多子图

plt.figure()
x=np.linspace(0,10,100)
plt.axes((0.1,0.5,0.8,0.4))
plt.plot(np.sin(x))# 绘制上子图
plt.axes((0.1,0.1,0.8,0.4))
plt.plot(np.cos(x))
```
